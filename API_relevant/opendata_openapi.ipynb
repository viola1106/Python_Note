```python
# 先import需要的套件
import googlemaps
import pandas as pd
import requests
```

```python
water_data = requests.get("https://twd.water.gov.taipei/opendata/wqb/wqb.asmx/GetQualityData")
# 註:範例說明是用API去get但我沒有API所以資料型別可能不一樣需要做主轉換
```

```python
water_data
```

```python
# 藉由.text可以直接看到整筆資料，不是json格式
water_data.text
```

```python
# xml轉換成json的過程要先將xml資料轉換成dict資料型別之後再轉乘json
import xmltodict
import json

# xml轉成dict
dict_water_data = xmltodict.parse(water_data.text)
# 再轉成json，這邊注意!!要加ensure_ascii = False，不然會變成utf8中文編碼會亂掉
json_water_data = json.dumps(dict_water_data, ensure_ascii = False)

# print(dict_water_data)
print(json_water_data)
```

```python
# 用DataFrame創立表格，然後往下層去找資料，這邊用dict資料型態就可以了
pd.DataFrame(dict_water_data["DataSet"]["diffgr:diffgram"]["NewDataSet"]["qua_data"])
```
