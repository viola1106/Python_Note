### Feature Engineering

```python
print("shape of before outlier:", train.shape)
train = train[np.abs(train['count'] - train['count'].mean()) <= (3*train['count'].std())]
print("shape of after outlier:", train.shape)
```

#### Data Preprocessing

```python
# 可以先把train和test合併再一起做，之後再分開
data = pd.concat([train, test], ignore_index=True)
data.head(5)
```

```python
# 處理時間
data['date'] = data.datetime.apply(lambda x : x.split()[0])
data['year'] = data.datetime.apply(lambda x : x.split()[0].split("-")[0])
data['hour'] = data.datetime.apply(lambda x : x.split()[1].split(":")[0]).astype("int")
data['weekday'] = data.date.apply(lambda datestring : datetime.strptime(datestring, "%Y-%m-%d").weekday())
data['month'] = data.date.apply(lambda datestring : datetime.strptime(datestring, "%Y-%m-%d").month)
```

```python
data.head(5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)  # 畫2*2的圖 -> 4個小圖
fig.set_size_inches(12, 10)  # 大小:長*寬:12*10

# sns.distplot()  # distribution:資料分布圖
# 分別畫4個圖:temp, atemp, humidity, windspeed
sns.set()
sns.histplot(data['temp'], ax=axes[0][0], kde=True, stat='density')
sns.histplot(data['atemp'], ax=axes[0][1], kde=True, stat='density')
sns.histplot(data['humidity'], ax=axes[1][0], kde=True, stat='density')
sns.histplot(data['windspeed'], ax=axes[1][1], kde=True, stat='density')

axes[0][0].set(xlabel='temp', title='distribution of temp')
axes[0][1].set(xlabel='atemp', title='distribution of atemp')
axes[1][0].set(xlabel='temp', title='distribution of humidity')
axes[1][1].set(xlabel='temp', title='distribution of windspeed')
```

windspeed應該是連續資料，但中間卻有空值，如果直接進入訓練，就會有很不準的模型，因此要先處裡空值問題。  
而依據圖形，Kaggle上有人猜測是因為風速測量不準確，所以沒有測量到的都被補值:0，所以0才會這麼多

```python
# 重新預測風速windspeed
# 先分成風速=0和風速!=0的資料
dataWind0 = data[data['windspeed'] == 0]
dataWindNot0 = data[data['windspeed'] != 0]
# 用RandomForestRegressor去預測風速是0的值，實際上有可能是多少
rfModel_wind = RandomForestRegressor(n_estimators=1000, random_state=42)  # n_estimators:1000樹做預測，random_state可以選任意值:隨機方式的設定
windcolumns = ['season', 'weather', 'humidity', 'month', 'temp', 'year', 'atemp']  # 預測風速的欄位
rfModel_wind.fit(dataWindNot0[windcolumns], dataWindNot0['windspeed'])  # train model:參數1:train的X資料，參數2:預測的結果
```

```python
# train完之後用predict()預測結果
wind0values = rfModel_wind.predict(X = dataWind0[windcolumns])
# 將預測出來的值代入原本風速值是0資料
dataWind0.loc[:, 'windspeed'] = wind0values
data = pd.concat([dataWindNot0, dataWind0], ignore_index=True)
data.head(10)
```

```python
sns.histplot(data['windspeed'], kde=True, stat='density')  # 預測後的風速圖
```

```python
# Feature Engineering結束後將資料重新分為train, test
dataTrain = data[pd.notnull(data['count'])].sort_values(by=['datetime'])
dataTest = data[~pd.notnull(data['count'])].sort_values(by=['datetime'])

datetimecol = dataTest['datetime']  # 最後submit時需要用到的資訊
yLabels = dataTrain['count']  # training data的y:腳踏車被租借的數量/小時
yLabelsLog = np.log(yLabels)  # 對y取log:原因如下
```

```python
sns.histplot(yLabels, kde=True, stat='density')  # 每小時租借數量的分布
```

圖形不是常態分佈，而是偏左的分佈。
在ML的model大多都是在常態分佈的假設上，所以要把它拉回成常態分佈，而其中一個方式就是: 取log


```python
sns.histplot(yLabelsLog, kde=True, stat='density') 
```

```python
# 丟掉不需要的欄位
dropfeatures = ['casual', 'count', 'datetime', 'date', 'registered']
dataTrain.drop(columns=dropfeatures, inplace=True)
dataTest.drop(columns=dropfeatures, inplace=True)
```

```python
# train model: RandomForestRegressor
rfModel = RandomForestRegressor(n_estimators=1000, random_state=42)
yLabelsLog = np.log(yLabels)
rfModel.fit(dataTrain, yLabelsLog)  # 要train的data:dataTrain, 取log的y
preds = rfModel.predict(X = dataTrain)  # 預測的值
```

```python
preds
```

```python
# 把資料放到submission:提交的資料是test
predsTest = rfModel.predict(X = dataTest)
submission = pd.DataFrame({
    'datetime': datetimecol,
    'count': [max(0, x) for x in np.exp(predsTest)]
})
submission.to_csv('bike_prediction_RFModel.csv', index=False)  # 不需要index
```
