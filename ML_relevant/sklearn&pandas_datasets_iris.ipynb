```python
from sklearn import datasets  # 從sklearn import datasets
```

```python
iris = datasets.load_iris()  # 在load iris這個資料集(iris:鳶尾花)
```

```python
iris.keys()  # load下來的資料集是dict型態，所以先看他所有的key
```

```python
# 查看各筆資料長什麼樣子
print(iris["DESCR"])  # 其中有一個'DESCR'是describe敘述，說明iris這個資料集
"""Number of Instances: 150 (50 in each of three classes):有150筆資料
   Attributes: 4 numeric: 有4個類型(花萼、花瓣的長寬)
   有3個種類的iris
   Summary Statistics:統計量表(Max, Min, Mean, etc.)
"""
```

```python
print(iris["data"])  # 資料是陣列的形式array但沒有column name
```

```python
print(iris["feature_names"])  # 可以看到feature_names這個資料裡面是data的資料名稱(欄位)
```

```python
print(iris["target"])
print(iris["target_names"])  # 再來看target的內容，可以同時看target_names，得知都是0、1、2，可以知道是種類(共3種)
```

```python
# 接著import pandas整理資料
import pandas as pd
```

```python
## 創建表格
# 創立iris["data"]的table，同時給予欄位名稱iris["feature_names"]
"""資料特徵用x表示"""
x = pd.DataFrame(iris["data"], columns = iris["feature_names"])
x  
```

```python
"""資料的結果即預測結果用y表示"""
# 創建target的表格(table)，即我們的預測結果
y = pd.DataFrame(iris["target"], columns = ["target"])  
# 欄位名稱:target，這邊注意不是target_names，因為那是種類名稱(有3個)，但這個table只有一個column
y
```

```python
# 試著合併x y兩個table用:pd.concat(注意合併方向)
data = pd.concat([x, y], axis=1)  # axis=1是橫向合併(欄位增加)
data
```

```python
"""種類0、1、2的表示法不夠直覺，因此將target的內容轉換成tarfet_names的內容，即將012轉換成種類名稱"""
# 首先用dict對應0、1、2&種類名稱
# 用enumerate()
# 用法:enumerate(iterable,start_index)前者是可跌代的對象(例list資料型態)，後者是起始編號(數字)，若沒有設定起始編號就由0開始
# 如果要設起始編號，寫法:enumerate(對象, start=數字)
print(dict(enumerate(iris["target_names"])))  # 因為target從0開始所以不設起始編號
```

```python
# 接著先轉換成花朵名稱類別
# 用map() function轉換
y1 = pd.DataFrame(pd.Series(iris["target"]).map(dict(enumerate(iris["target_names"]))), columns = ["target_names"])
# DataFrame做成表格table，只有一欄資料轉換用Series
# 之後用 .map() 對每一個物件元素iterable做處理(enumerate())
y1
```

```python
# 再合併一次x y1，完成表格處理
data = pd.concat([x, y1], axis=1)
data
```
