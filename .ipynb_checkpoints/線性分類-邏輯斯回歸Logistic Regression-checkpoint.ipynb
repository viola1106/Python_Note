```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # 畫圖用
import seaborn as sns  # 畫圖用
```

```python
%matplotlib inline
```

```python
# 匯入iris資料集
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'], columns = iris['feature_names'])
print("target_names:" + str(iris['target_names']))
y = pd.DataFrame(iris['target'], columns = ['target'])
iris_data = pd.concat([x, y], axis=1)
iris_data.head(3)
```

```python
# 通常在建立model時會用train_test_split把資料分為七三筆，七是train model剩下的三來看model好不好
from sklearn.model_selection import train_test_split
```

```python
# 把資料切割x, y的train, test
# 使用的特徵是: 花萼的長度&花瓣長度
x_train, x_test, y_train, y_test = train_test_split(
    iris_data[['sepal length (cm)', 'petal length (cm)']], iris_data[['target']], test_size=0.3, random_state=0)
```

```python
# random_state=0:隨機找資料做training，可以看到x_train的index是隨機的
print("x_train筆數:" + str(len(x_train)))
x_train
```

```python
# 前面70%筆資料做train，剩下的30%可以看這個model的正確率是多少，可以知道未來的預測正確率大概會是多少
print("x_test筆數:" + str(len(x_test)))
x_test
```

```python
"""
在做Logistic Regression的時候要打特徵範圍(sacle)縮小到一樣的範圍，可以都縮小0~1之間，
或是用StandardScaler，把資料標準化
"""

# 資料標準畫:值減掉平均，除上標準差
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
# sigmoid函式圖型

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.plot(z, phi_z)
plt.axvline(0.0, color = 'k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# 
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
```
