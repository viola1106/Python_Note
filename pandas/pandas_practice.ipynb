```python
# 先import
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
# 用跟之前一樣的dataset_iris
iris = datasets.load_iris()
x = pd.DataFrame(iris["data"], columns = iris["feature_names"])  # x表格是iris_data
print("target_names:" + str(iris["target_names"]))  # target是iris的分類，共3類
y = pd.DataFrame(iris["target"], columns = ["target"])  # y表格是分類
data = pd.concat([x, y], axis=1)  # 合併x,y兩個表格，方向是橫向的(欄位增加)
data.head(3)  # 看前3筆資料確認
# 可以看到下方table最後一欄是target也就是原本的y表格
```

```python
"""
pandas的背後資料結構都是numpy的array，而不用python內建的list，
原因是因為python可以容許list裡面有不同的資料型態，
比如說在一個list裡面可以同時存在str int float等等，這會造成他在運算的時候速度與記憶體的效能都不好，
所以pandas背後全部都是用numpy的array做處理
"""
list1 = ["1", 2, 3.5, 4]
print(list1)
np_list1 = np.array(list1)
print(np_list1)
# 可以看到用原本的python_list就會是原本的資料型態
# 但如果用np.array轉印出來會統一處理成str型態
```

### Data Grouping (資料分群)

```python
data.head()  # 回顧一下資料
```

```python
# 每一個target群體的花萼長是多少、寬是多少
data.groupby(by = "target").sum()
```

```python
# 如果想看平均值的話:.mean()
data.groupby(by = "target").mean()
```

```python
# 如果想知道每一個group的內容的話就用for loop
for i in data.groupby(by = "target"):
    print(i)
```

```python
# 如果是自訂條件不用內建函式(sum,mean)，就用apply
# 花萼長度第二大的資料
# 用lambda來執行:先進行花萼長度排序
# ascending=False是倒序，由大排到小
# [1:2]是取第二筆資料，注意index起始值是0
data.groupby(by = "target").apply(lambda x: x.sort_values(by = ["sepal length (cm)"], ascending=False)[1:2])
# 顯示各分群target第二大的花萼長度
```

```python
# 確認函式執行成果是否正確:target 0的第二大花萼長度是否為5.7
# 先篩選target==0，再選出花萼長度欄位，在排序大小
data[data['target'] == 0]['sepal length (cm)'].sort_values()
```

可以看到花萼長度最大的是5.8，第二大的是5.7，所以自訂函式的執行結果是正確的
