```python
# 試著找出全臺灣"寵物"關鍵字的店家:找出店家id，就可以從id output店家資訊
# 先創立全台縣市的list
cities = ["基隆市", "臺北市", "新北市", "桃園市", "新竹縣", "苗栗縣", "臺中市","彰化縣", "南投縣",
          "雲林縣", "嘉義縣","臺南市", "高雄市", "屏東縣", "宜蘭縣", "花蓮縣", "台東縣", "澎湖縣"]
# 存放搜尋結果id的list
result_ids = []

# 搜尋全台寵物店家
for city in cities:
    geocode_result = gmaps.geocode(city)  # 找出所有縣市的經緯度
    loc = geocode_result[0]["geometry"]["location"]
    # 找出半徑10000公尺的店家並extend到list
    result = []
    query_result = gmaps.places_nearby(keyword="寵物", type="store", location=loc, radius=10000)
    result.extend(query_result["results"])
    # 接著處理next_page_token:增加一個參數page_token
    while query_result.get("next_page_token"):
        time.sleep(3)
        query_result = gmaps.places_nearby(page_token=query_result["next_page_token"])
        result.extend(query_result["results"])
    print(city + "半徑10000公尺內的寵物店家數(google map api上限60家):" + str(len(result)))
    # 把所有的id都存在list裡，之後查資料用
    for place in result:
        result_ids.append(place["place_id"])
```

```python
gmaps.place(place_id=result_ids[-300], language="zh-TW")["result"]
```

```python
store_info = []
# 去除重複的id
result_ids = list(set(result_ids))
# 將店家資訊放進一個list裡
for ids in result_ids:
    store_info.append(gmaps.place(place_id=ids, language="zh-TW")["result"])
    # 確認append進度，每100筆資料就印出來
    if len(store_info)%100 == 0:
        print(len(store_info))
```

```python
output = pd.DataFrame.from_dict(store_info)
```
